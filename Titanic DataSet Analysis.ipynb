```python
import numpy as np
import pandas as pd
import random as rnd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression,SGDClassifier,Perceptron
from sklearn.svm import SVC,LinearSVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
```

```python
train=pd.read_csv("train.csv")         # Load the data set
test=pd.read_csv("test.csv")
train.info()                           # check for null and dtype
test.info()
```

```python
train.columns
```

```python
train[['Pclass','Survived']].groupby(['Pclass'],as_index=False).mean().sort_values(['Survived'],ascending=False)  # Survived by Pclass
```

```python
train[['Sex','Survived']].groupby(['Sex'],as_index=False).mean().sort_values(['Survived'],ascending=False)  # Survived by Sex
```

```python
train[['SibSp','Survived']].groupby(['SibSp'],as_index=False).mean().sort_values(['Survived'],ascending=False)  # Survived by SibSp
```

```python
train[['Parch','Survived']].groupby(['Parch'],as_index=False).mean().sort_values(['Survived'],ascending=False)  # Survived by Parch
```

```python
train[['Embarked','Survived']].groupby(['Embarked'],as_index=False).mean().sort_values(['Survived'],ascending=False)  # Survived by Embarked
```

```python
data=pd.concat([train,test],ignore_index=True)         #join train and test to get a single dataset
data.info()
```

```python
label_encoder=LabelEncoder()

data['Embarked']=label_encoder.fit_transform(data['Embarked'])
data['Sex']=label_encoder.fit_transform(data['Sex'])
```

```python

train['Age'].fillna(train['Age'].mean(), inplace=True)
train['Embarked'].fillna(train['Embarked'].mode()[0], inplace=True)
test['Age'].fillna(test['Age'].mean(), inplace=True)
test['Embarked'].fillna(test['Embarked'].mode()[0], inplace=True)
```

```python
Y_train=train['Survived']
X_train=train[['Pclass', 'Sex', 'Age', 'SibSp','Parch','Embarked']]

X_test=test[['Pclass', 'Sex', 'Age', 'SibSp','Parch','Embarked']]
```

```python
label_encoder=LabelEncoder()
X_train['Sex']=label_encoder.fit_transform(X_train['Sex'])

X_train['Embarked']=label_encoder.fit_transform(X_train['Embarked'])

X_test['Sex']=label_encoder.fit_transform(X_test['Sex'])

X_test['Embarked']=label_encoder.fit_transform(X_test['Embarked'])
```

```python
X_train.head()
```

```python
logistic_model=LogisticRegression()
logistic_model.fit(X_train,Y_train)
```

```python
Y_predicted=logistic_model.predict(X_test)
accuracy_logistic=round(logistic_model.score(X_train,Y_train)*100,2)
accuracy_logistic
```

```python
PredictedResult=pd.DataFrame({'PassengerId':test.index+1,'Survived':Y_predicted})
```

```python
PredictedResult
```

```python
svc_model=SVC()
svc_model.fit(X_train,Y_train)
Y_predict=svc_model.predict(X_test)
accuracy_svc=round(svc_model.score(X_train,Y_train)*100,2)
accuracy_svc
```

```python
rf_model=RandomForestClassifier()
rf_model.fit(X_train,Y_train)
Y_predict=rf_model.predict(X_test)
accuracy_rf=round(rf_model.score(X_train,Y_train)*100,2)
accuracy_rf
```

```python
knn_model=KNeighborsClassifier()
knn_model.fit(X_train,Y_train)
Y_predict=knn_model.predict(X_test)
accuracy_knn=round(knn_model.score(X_train,Y_train)*100,2)
accuracy_knn
```

```python
nb_model=GaussianNB()
nb_model.fit(X_train,Y_train)
Y_predict=nb_model.predict(X_test)
accuracy_nb=round(nb_model.score(X_train,Y_train)*100,2)
accuracy_nb
```

```python
p_model=Perceptron()
p_model.fit(X_train,Y_train)
Y_predict=p_model.predict(X_test)
accuracy_p=round(p_model.score(X_train,Y_train)*100,2)
accuracy_p
```

```python
sgd_model=SGDClassifier()
sgd_model.fit(X_train,Y_train)
Y_predict=sgd_model.predict(X_test)
accuracy_sgd=round(sgd_model.score(X_train,Y_train)*100,2)
accuracy_sgd
```

```python
dt_model=DecisionTreeClassifier()
dt_model.fit(X_train,Y_train)
Y_predict=dt_model.predict(X_test)
accuracy_dt=round(dt_model.score(X_train,Y_train)*100,2)
accuracy_dt
```

```python
model=pd.DataFrame({'Model':['Logistic Regression','Support Vector Machine','Random Forest','K Nearest Neighbors','Gaussian Naive Bayes','Perceptron','Stochastic Gradient Descent Classifier','Decision Tree'],'Score':[accuracy_logistic,accuracy_svc,accuracy_rf,accuracy_knn,accuracy_nb,accuracy_p,accuracy_sgd,accuracy_dt]})
model.sort_values(by='Score',ascending=False)
```
